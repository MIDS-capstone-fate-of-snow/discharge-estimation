```python

import datetime
import pandas as pd


satellite = [
    "ECMWF/ERA5_LAND/HOURLY",
    "MODIS/006/MOD16A2",
]
band = [
    "total_precipitation",
    "ET",
]
aggs = [
    "mean",
    None,
]
gage = [
    "11266500",
    "11402000",
] 
start_date = [
    datetime.datetime(2010, 1, 1),
    datetime.datetime(2010, 1, 1),
]
end_date = [
    datetime.datetime(2020, 12, 31),
    datetime.datetime(2020, 12, 31),
]
time_s_per_year = [
    4750.38,
    624.21
]


df = pd.DataFrame({
    "satellite": satellite,
    "band": band,
    "aggs": aggs,
    "gage": gage,
    "start_date": start_date,
    "end_date": end_date,
    "time_s_per_year": time_s_per_year,
})
df["days"] = (df["end_date"] - df["start_date"]).dt.days
df["est_total_runtime_s"] = df["days"] / 365.25 * df["time_s_per_year"]
df["est_total_runtime_m"] = df["est_total_runtime_s"] / 60
df["est_total_runtime_h"] = df["est_total_runtime_m"] / 60
df
```
